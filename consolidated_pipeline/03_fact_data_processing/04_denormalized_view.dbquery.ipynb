```sql

create or replace view fmcg.gold.vw_fact_orders as (
select
 fo.date as order_date,
 fo.product_code,
 fo.customer_code,

 dd.date,
 dd.year,
 dd.month,
 dd.month_name,
 dd.quarter,

 dc.customer,
 dc.market,
 dc.platform,
 dc.channel,

 dp.division,
 dp.category,
 dp.product,
 dp.variant,

 fo.sold_quantity,
 gp.price_inr,

 (fo.sold_quantity*gp.price_inr) as total_amount_inr
 
 from fmcg.gold.fact_orders fo 
 
 --date table
 left join
 fmcg.gold.dim_date dd
 on fo.date=dd.date


 left join 
 fmcg.gold.dim_customers dc
 on fo.customer_code=dc.customer_code

 left join
  fmcg.gold.dim_products dp
  on fo.product_code=dp.product_code

  left join 
   fmcg.gold.dim_gross_price gp
   on fo.product_code=gp.product_code
   and year(fo.date)=gp.year);
   
   select * from fmcg.gold.vw_fact_orders;






```
